```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Results for 2 host setup (clients reside in one host and all servers reside in the other host)
# The server is a 2016 MacBook Pro, and the client is a 2015 MacBook Pro
x_num_clients = [1, 2, 4, 8, 16, 32, 64, 128]
avg_pos_30nodes = [22.19, 25.43, 20.53, 26.42, 25.16, 26.99, 29.36, 41.12]
med_pos_30nodes = [16, 21, 14, 18, 18, 17, 19, 25]

avg_pos_10nodes = [6.76, 7.38, 6.5, 8.62, 8.839, 11.13, 12.65, 22.22]
med_pos_10nodes = [5, 5, 4, 5, 5, 6, 7, 13]

avg_pos_3nodes = [1.545, 1.618, 2.714, 2.636, 3.663, 4.324, 8.858, 14.79]
med_pos_3nodes = [0, 1, 1, 1, 2, 2, 4, 7]
```

```python
plt.plot(x_num_clients, avg_pos_30nodes, 'r-', label="Avg 30 Server Nodes")
plt.plot(x_num_clients, avg_pos_10nodes, 'b-', label="Avg 10 Server Nodes")
plt.plot(x_num_clients, avg_pos_3nodes, 'g-', label="Avg 3 Server Nodes")

plt.plot(x_num_clients, med_pos_30nodes, 'r--', label="Median 30 Server Nodes")
plt.plot(x_num_clients, med_pos_10nodes, 'b--', label="Median 10 Server Nodes")
plt.plot(x_num_clients, med_pos_3nodes, 'g--', label="Median 3 Server Nodes")
plt.legend()
plt.title("Client Server Distributed Priority Queue Performance")
plt.xlabel("Number of Concurrent Clients")
plt.ylabel("Index of Popped Element")
plt.rcParams['figure.figsize'] = [10, 50]
plt.show()
```
